## Content: PLOT figure S1 - land-sea-fraction matrix

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
OMs = ["guyane","guadeloupe","martinique","reunion","nc","polynesie"]

Stations={"guyane":["Cayenne-Matoury","Maripasoula","Saint-Georges","Saint-Laurent"],
          "guadeloupe" : ["Saint-Barthelemy","Raizet"],"martinique":["Lamentin"],"reunion":["Pamandzi","Gillot","Tromelin"],
          "nc":["Koumac","Tontouta","Noumea","Ouanaham","Maopoopo","Hihifo"],
          "polynesie":["Faaa","Hiva-oa","Mangareva","Rapa","Takaroa"]}

Modeles=["ACCESS-CM2","CanESM5","CNRM-ESM2-1","CNRM-CM6-1","CNRM-CM6-1-HR",
         "EC-Earth3","EC-Earth3-CC","EC-Earth3-Veg-LR","KACE-1-0-G",
         "IPSL-CM6A-LR","MIROC6","MIROC-ES2L","MPI-ESM1-2-HR",
         "MPI-ESM1-2-LR","NorESM2-MM"]
```

### Create a dataframe

```python
lsf_cmip=pd.DataFrame(index=Modeles)


for OM in OMs:
    for station in Stations[OM]:
        data_cmip=pd.read_csv("/CMIP6/LSF/lsf_cmip6_"+station+".csv")
        lsf_cmip[station]=data_cmip[station].values
        
lsf_cmip
```

### PLOT

```python
xlabels=["Cayenne-Matoury","Maripasoula","Saint-Georges","Saint-Laurent","Saint-Barthelemy","Le Raizet","Lamentin",
         "Pamandzi","Gillot","Tromelin","Koumac",
        "La Tontouta","Noumea","Ouanaham","Maopoopo","Hihifo",
         "Faaa","Hiva-oa","Mangareva","Rapa","Takaroa"]

ylabels=Modeles

for tab in [lsf_cmip]:
    fig, ax = plt.subplots(ncols=1,figsize=(15,11))
    sns.heatmap(tab, vmin=0, vmax=100, cmap = 'BrBG_r', linewidth=0.5,annot=True, annot_kws={'fontsize':10},cbar_kws={},fmt=".1f")
    ax.xaxis.tick_top()
    ax.set_xticklabels(xlabels, rotation=45,ha='left')
    ax.set_yticklabels(ylabels, rotation=0)
    ax.set_title("Land-sea fraction - CMIP6",fontsize=16)
    
    # saving
    plt.savefig('/figures/lsf_cmip6_per_station.png',transparent=False,dpi=300)
    plt.savefig('/figures/lsf_cmip6_per_station.svg',transparent=False,dpi=300)
```
